#### 3. Normalize the data

Returns the data to the range (0,1).

```python
train_X = x_train.astype("float") / 255.0
valid_X = x_valid.astype("float") / 255.0
test_X = x_test.astype("float") / 255.0
```

Labels numbered from 0 to 42 are equivalent to the list above.

Switch to one hot form.

One hot is just a way to transfer labels to ensure the similarity of the labels with each other. The image below illustrates the onehot matrix.

onehot.jpeg

```python
Y_train = to_categorical(y_train,num_classes=len(classNames))
Y_valid = to_categorical(y_valid,num_classes=len(classNames))
Y_test = to_categorical(y_test,num_classes=len(classNames))
```

#### 4. Model design

We do the convolutions, max-pooling and then feed the half fully-connected network and reduce the layer dimension to exactly the number of classes.

Best accuracy: 93.39% within 30 epochs

Best accuracy: 96.65% with class_weight within 30 epochs

Because of imbalanced data, accuracy does not clearly represent the performance of the model, instead AUC and F1_Score will evaluate the classifier better.

See more: [Performance metrics for imbalanced multiclass dataset](https://stats.stackexchange.com/questions/463224/which-performance-metrics-for-highly-imbalanced-multiclass-dataset)



```python
learning_rate = 1e-2
batch_size = 64
num_epochs = 30
```

```python
width = 32
height = 32
classes = 43
```

```python
shape = (width,height, 3)
```

```python
model = Sequential()

model.add(Conv2D(32, (3 , 3), padding='same', input_shape=shape))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(Conv2D(32, (3 , 3), padding='same'))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3 , 3), padding='same'))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(Conv2D(64, (3 , 3), padding='same'))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(512))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(classes))

model.add(Activation("softmax"))
```

Check the structure of the model before training.

```python
model.summary()
```

#### 5. Apply class weight and data agumentation before training

We can rotate, zoom, resize for more features while keeping the same label

```python
aug = ImageDataGenerator(rotation_range=18, zoom_range=0.15, width_shift_range=0.2, height_shift_range=0.2, shear_range=0.15,
                                                                                           horizontal_flip=True, fill_mode="nearest")
```

```python
opt = SGD(learning_rate=learning_rate, momentum=0.9)
model.compile(loss="categorical_crossentropy", optimizer=opt, metrics=(["accuracy"], tf.keras.metrics.Precision(name='precision'),
                                                                                                        tf.keras.metrics.Recall(name='recall'),
                                                                                                        tfa.metrics.F1Score(num_classes=len(classNames),
                                                                                                                            average='macro',
                                                                                                                            threshold=0.5,
                                                                                                                            name='F1_score'),
                                                                                                        tf.keras.metrics.AUC(name='AUC')))
```

```python

checkpoint = ModelCheckpoint("model_weights.h5", monitor='val_F1_score',
                             save_weights_only=True, mode='max', verbose=1)
early_stopping = EarlyStopping(monitor='val_accuracy',
    min_delta=0.00005,
    patience=11,
    verbose=1,
    restore_best_weights=True,
)

lr_scheduler = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.5,
    patience=7,
    min_lr=0.00001,
    verbose=1,
)

callbacks = [
    PlotLossesCallback(),
    early_stopping,
    lr_scheduler,
    checkpoint
]
```

```python
class_weights = class_weight.compute_class_weight(class_weight = "balanced", classes= np.unique(y_train), y= y_train)
class_weight = dict(enumerate(class_weights.flatten()))
print(class_weight)
```

```python
print("Start Training...")
H = model.fit(aug.flow(x_train, Y_train, batch_size=batch_size), validation_data=(x_test, Y_test), 
                        steps_per_epoch=train_X.shape[0] // batch_size, epochs=num_epochs, verbose=1, callbacks = callbacks, class_weight = class_weight)
```
